```python
import pandas as pd
import numpy as np
import pickle
```

```python
df_train=pd.read_csv('train_final.csv',index_col=False)
labels=df_train[['784']]
```

```python
df_train.drop(df_train.columns[[784]],axis=1,inplace=True)
df_train.head()
```

```python
np.random.seed(1212)
import keras
from keras.models import Model
from keras.layers import *
from keras import optimizers
from keras.layers import Input, Dense
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils
```

```python
labels=np.array(labels)
```

```python
from keras.utils.np_utils import to_categorical
cat=to_categorical(labels,num_classes=22)
```

```python
print(cat[0])
```

```python
df_train.head()
```

```python
l=[]
for i in range(24533):
    l.append(np.array(df_train[i:i+1]).reshape(28,28,1))
```

```python
np.random.seed(7)
```

```python
model=Sequential()
model.add(Conv2D(30, (5, 5), input_shape=( 28, 28,1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(15, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(206, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(22, activation='softmax'))
```

```python
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(np.array(l), cat, epochs=10, shuffle=True)
```

```python
model.save('model1.keras')
```

```python
import math
def trigo(sin,cos,tan,s):
    if sin==1:
        s=s[3:]
        s1=math.sin(float(s))
        
    elif cos==1:
        s=s[3:]
        s1=math.cos(float(s))
        
    elif tan==1:
        s=s[3:]
        s1=math.tan(float(s))
    return(s1)
```

```python
def prediction(train_data):
    import math
    s=''
    sin=0
    cos=0
    tan=0
    para1=0
    para2=0
    eqq=0
    for i in range(len(train_data)):
        train_data[i]=np.array(train_data[i])
        train_data[i]=train_data[i].reshape(1,28,28,1)
        result=model2.predict_classes(train_data[i])
        if(result[0]==10):
            s=s+' -'
        if(result[0]==11):
            s=s+' +'
        if(result[0]==12):
            s=s+'*'
        if(result[0]==0):
            s=s+'0'
        if(result[0]==1):
            s=s+'1'
        if(result[0]==2):
            s=s+'2'
        if(result[0]==3):
            s=s+'3'
        if(result[0]==4):
            s=s+'4'
        if(result[0]==5):
            s=s+'5'
        if(result[0]==6):
            s=s+'6'
        if(result[0]==7):
            s=s+' 7'
        if(result[0]==8):
            s=s+'8'
        if(result[0]==9):
            s=s+'9'
        if(result[0]==13):
            s=s+'/'
        if(result[0]==14):
            s=s+'('
        if(result[0]==15):
            s=s+')'
        if(result[0]==16):
            s=s+'sin'
            sin=1
            
        if(result[0]==17):
            s=s+'cos'
            cos=1
        if(result[0]==18):
            s=s+'tan'
            tan=1
        if(result[0]==19):
            s=s+'a' 
            eqq=1
        #if(result[0]==20):
            #s=s+'='
        if(result[0]==21):
            s=s+' b'


    print('The given input string is'+'\n')
    print(s +'\n')
    if sin==1:
        s1=trigo(sin,cos,tan,s)
        return s,s1
    elif cos==1:
        s1=trigo(sin,cos,tan,s)
        return s,s1
    elif tan==1:
        s1=trigo(sin,cos,tan,s)
        return s,s1
    else:
        if eqq!=1:
                
            print('Final Answer'+'\n')
                
    return (s)
```

```python


import cv2
#"C:/Users/abhiruchi/Pictures/test57.jpg"
import numpy as np
def imginp(r):
    print(r)
    img = cv2.imread(r,cv2.IMREAD_GRAYSCALE)
    
    if img is not None:
        
        img=~img
        ret,thresh=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
        ctrs,ret=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
        cnt=sorted(ctrs, key=lambda ctr: cv2.boundingRect(ctr)[0])
        w=int(28)
        h=int(28)
        train_data=[]
        #print(len(cnt))
        rects=[]
        for c in cnt :
            x,y,w,h= cv2.boundingRect(c)
            rect=[x,y,w,h]
            rects.append(rect)
        #print(rects)
        bool_rect=[]
        for r in rects:
            l=[]
            for rec in rects:
                flag=0
                if rec!=r:
                    if r[0]<(rec[0]+rec[2]+10) and rec[0]<(r[0]+r[2]+10) and r[1]<(rec[1]+rec[3]+10) and rec[1]<(r[1]+r[3]+10):
                        flag=1
                    l.append(flag)
                if rec==r:
                    l.append(0)
            bool_rect.append(l)
        #print(bool_rect)
        dump_rect=[]
        for i in range(0,len(cnt)):
            for j in range(0,len(cnt)):
                if bool_rect[i][j]==1:
                    area1=rects[i][2]*rects[i][3]
                    area2=rects[j][2]*rects[j][3]
                    if(area1==min(area1,area2)):
                        dump_rect.append(rects[i])
        #print(len(dump_rect)) 
        final_rect=[i for i in rects if i not in dump_rect]
        #print(final_rect)
        for r in final_rect:
            x=r[0]
            y=r[1]
            w=r[2]
            h=r[3]
            im_crop =thresh[y:y+h+10,x:x+w+10]


            im_resize = cv2.resize(im_crop,(28,28))
            

            im_resize=np.reshape(im_resize,(28,28,1))
            train_data.append(im_resize)
        return prediction(train_data)
        
```

```python
def eqsol(val,val1):
    import mpmath
    import sympy
    from sympy import symbols, Eq, solve
    a, b = symbols('a b')
    eq1 = Eq(eval(val))
    print(eq1)
    eq2 = Eq(eval(val1))
    print(eq2)
    return solve((eq1,eq2), (a, b))
```

```python
from flask import *
import numpy as np
import keras
import cv2
model2=keras.models.load_model('model1.keras')
app = Flask(__name__)
r="abc"
app.config['Image_folder']='static/images/'
@app.route("/normal.html", methods=["POST", "GET"])
def upload():
    global r
    
    if request.method == 'POST':
        
        f = request.files['file']
        f.save(f.filename)
        img1=cv2.imread(f.filename)
        r = f.filename
        direct=app.config['Image_folder']+str(f.filename)
        cv2.imwrite(direct,img1)
        
        b=imginp(r)
        return render_template("normal.html", name=f.filename,ans1=b,ans=eval(b),imgg=direct)
    return render_template("normal.html")

@app.route("/trig.html", methods=["POST", "GET"])
def trigno():
    global r
    
    if request.method == 'POST':
        
        f = request.files['file']
        f.save(f.filename)
        img2=cv2.imread(f.filename)
        r = f.filename
        direct=app.config['Image_folder']+str(f.filename)
        cv2.imwrite(direct,img2)
        b,d=imginp(r)
        return render_template("trig.html", name=f.filename,ans1=b,ans=(d),imgg=direct)
    return render_template("trig.html")

@app.route("/simul.html", methods=["POST", "GET"])
def success():
    global r
    if request.method == 'POST':
        f = request.files['file1']
        f2 = request.files['file2']
        f.save(f.filename)
        f2.save(f2.filename)
        img3=cv2.imread(f.filename)
        img4=cv2.imread(f2.filename)
        r = f.filename
        r2 = f2.filename
        direct=app.config['Image_folder']+str(f.filename)
        cv2.imwrite(direct,img3)
        direct2=app.config['Image_folder']+str(f2.filename)
        cv2.imwrite(direct2,img4)
        
        b=imginp(r)
        d=imginp(r2)
        result=eqsol(b,d)
        return render_template("simul.html", name1=f.filename,name2=f2.filename,ansb=b,ansd=d,ans=result,imgg1=direct,imgg2=direct2)
    return render_template("simul.html")
    
@app.route('/index.html',methods=['GET','POST'])
def index():
    if request.method == 'POST':
        if request.form.get('normal'):
            return redirect(url_for('upload'))
        if request.form.get('trig'):
            return redirect(url_for('trigno'))
        if request.form.get('simul'):
            return redirect(url_for('success'))
    return render_template('index.html')

if __name__ == '__main__':
    app.run(debug=False)
    
#print(r)
```
